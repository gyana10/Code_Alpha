#REPLACING THE BLANKS WITH "0" AND COVERTING THE DATATYPE TO FLOAT

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["MonthlyCharges"]=df["MonthlyCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
df["MonthlyCharges"]=df["MonthlyCharges"].astype("float")
```

#WE ARE CONVERTING 0 AND 1 VALUE TO YES AND NO TO MAKE IT UNDERSTANDABLE

```python
def convo(value):
    if value==1:
        return "yes"
    else:
        return "no"
df['SeniorCitizen']=df['SeniorCitizen'].apply(convo)
```

#NOW WE HAVE TO EXPLORE THE REASON FOR THE CHURN

```python
ax=sns.countplot(x='SeniorCitizen', data=df)
ax.bar_label(ax.containers[0])
plt.title("No. of customer by Senior Citizens")
plt.show()
```

```python
grouped = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='Count')
# Pivot so that each SeniorCitizen category is a row and columns correspond to Churn values.
pivot_df = grouped.pivot(index='SeniorCitizen', columns='Churn', values='Count').fillna(0)
# Ensure columns are in a known order (for example, "No" then "Yes")
pivot_df = pivot_df[['No', 'Yes']]
# Step 3: Compute percentages
# Compute the total count for each SeniorCitizen group
pivot_df['Total'] = pivot_df.sum(axis=1)
# Calculate percentages for each Churn value
pivot_df['No_perc'] = pivot_df['No'] / pivot_df['Total'] * 100
pivot_df['Yes_perc'] = pivot_df['Yes'] / pivot_df['Total'] * 100
# Step 4: Plot the stacked bar chart
plt.figure(figsize=(6, 4))
# First, plot the "No" segment
bar_no = plt.bar(pivot_df.index, pivot_df['No'], label='No', color='skyblue')
# Then, plot the "Yes" segment on top of "No"
bar_yes = plt.bar(pivot_df.index, pivot_df['Yes'], bottom=pivot_df['No'], label='Yes', color='salmon')
plt.xlabel("SeniorCitizen")
plt.ylabel("Count")
plt.title("Churn by SeniorCitizen (Stacked Bar Chart with % labels)")
# Step 5: Annotate with percentage labels
for idx in pivot_df.index:
    # Retrieve heights for each segment
    no_height = pivot_df.loc[idx, 'No']
    yes_height = pivot_df.loc[idx, 'Yes']
    
    # Retrieve computed percentages
    no_perc = pivot_df.loc[idx, 'No_perc']
    yes_perc = pivot_df.loc[idx, 'Yes_perc']
    
    # Annotate the "No" segment (if non-zero)
    if no_height > 0:
        plt.text(idx, no_height / 2, f"{no_perc:.1f}%", ha='center', va='center', color='black', fontsize=10)
    
    # Annotate the "Yes" segment (if non-zero)
    if yes_height > 0:
        plt.text(idx, no_height + yes_height / 2, f"{yes_perc:.1f}%", ha='center', va='center', color='black', fontsize=10)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
categorical_columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the figure and axes for subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 10))  # 3x3 grid

# Flatten axes array for easy iteration
axes = axes.flatten()

# Loop through each categorical column and create a countplot
for i, col in enumerate(categorical_columns):
    sns.countplot(data=df, x=col, ax=axes[i], palette="Set2",hue="Churn")
    axes[i].set_title(f"Count of {col}")  # Set title
    axes[i].set_xlabel("")  # Remove x-axis label
    axes[i].set_ylabel("Count")  # Set y-axis label
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-ticks for better visibility

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

The subplots display countplots of various service subscriptions (e.g., PhoneService, InternetService, StreamingTV) categorized by Churn status (Yes/No). Each plot highlights the distribution of customers who have or do not have a particular service and their corresponding churn rates. The visualizations indicate potential patterns in customer churn, such as higher churn rates among Fiber Optic internet users and those lacking online security or tech support. These insights can help identify key factors influencing customer retention.
